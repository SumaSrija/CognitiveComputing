Lab2-LinearRegression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate random data
np.random.seed(82)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Make predictions on new data
X_new = np.array([[0], [2]])
y_pred = model.predict(X_new)

# Visualize the results
plt.scatter(X, y, label='Training Data')
plt.plot(X_new, y_pred, 'r-', label='Linear Regression Predictions')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title('Linear Regression Example')
plt.show()
```

```python
from textblob import TextBlob

text = "This movie is just okay."

# Create a TextBlob object
blob = TextBlob(text)

# Analyze sentiment
sentiment = blob.sentiment

print(f"Sentiment: {sentiment.polarity}")  # Output: -0.8

# Classify sentiment
if sentiment.polarity > 0:
    print("Positive")
elif sentiment.polarity == 0:
    print("Neutral")
else:
    print("Negative")
```

lab 1 ANN model

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

# Step 1: Load the MNIST dataset from a CSV file
# Make sure your CSV file has columns for labels and pixel values (pixel1, pixel2, ..., pixel784)
mnist_data = pd.read_csv('/content/fashion-mnist_test.csv')

# Extract labels and features
labels = mnist_data['label'].values
features = mnist_data.drop('label', axis=1).values

# Reshape features to 28x28 images
features = features.reshape((-1, 28, 28, 1)).astype('float32') / 255.0

# One-hot encode labels
labels = to_categorical(labels)

# Split the data into training and testing sets
train_images, test_images, train_labels, test_labels = train_test_split(features, labels, test_size=0.2, random_state=42)

# Step 2: Build the neural network architecture
model = models.Sequential()
model.add(layers.Flatten(input_shape=(28, 28, 1)))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

# Step 3: Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Step 4: Train the model
model.fit(train_images, train_labels, epochs=7, batch_size=64, validation_split=0.2)

# Step 5: Evaluate the model on the test set
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'Test accuracy: {test_acc}')

# Optional: Save the model for later use
model.save('mnist_model.h5')
```

ex4-twitter data analysis - NLTK Library

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

nltk.download('punkt')
nltk.download('stopwords')

def preprocess_tweet(tweet):
    # Tokenize the tweet
    words = word_tokenize(tweet)

    # Remove stop words and punctuation
    stop_words = set(stopwords.words('english') + list(string.punctuation))
    filtered_words = [word.lower() for word in words if word.lower() not in stop_words]

    return ' '.join(filtered_words)

def analyze_sentiment(tweet):
    sid = SentimentIntensityAnalyzer()
    sentiment_score = sid.polarity_scores(tweet)['compound']

    # Classify sentiment based on the compound score
    if sentiment_score >= 0.05:
        return 'Positive'
    elif sentiment_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

if __name__ == "__main__":
    # Example tweets (replace with your own dataset or scraping logic)
    tweets = [
        "I love Python! It's an amazing language.",
        "This movie was terrible. I can't believe I wasted my time.",
        "Just finished reading a great book. Highly recommended!",
    ]

    print("Analyzing sentiment for tweets:\n")

    for tweet in tweets:
        preprocessed_tweet = preprocess_tweet(tweet)
        sentiment = analyze_sentiment(preprocessed_tweet)

        print(f"Original Tweet: {tweet}")
        print(f"Preprocessed Tweet: {preprocessed_tweet}")
        print(f"Sentiment: {sentiment}\n")

```
